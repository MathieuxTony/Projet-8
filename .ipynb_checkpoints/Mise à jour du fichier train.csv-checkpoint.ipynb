# Mise à jour du fichier train.csv

Le fichier train.csv téléchargé sur kaggle comporte des erreurs dont les corrections ont été données dans un fichier séparé train_updates.csv. Dans ce notebook nous allons simplement créer un nouveau fichier contenant les données d'entraînements corrigées.

```python
import pandas as pd
import missingno as mso
```

```python
train = pd.read_csv("Data/train.csv") 
train_updates = pd.read_csv("Data/train_updates.csv")
```

```python
print("train shape : ", train.shape)
print("train_updates shape : ", train_updates.shape)
```

```python
train.head()
```

```python
train_updates.head()
```

Il y a manifestement beaucoup de valeurs manquantes. Visualisons ça avec missingno.

```python
mso.bar(train_updates.sample(train_updates.shape[0]))
```

Les 'seq_id' qui nous sont fournis sans aucune autre donnée sont des identifiants à supprimer dans le fichier train.csv. Commençons par ça.

```python
drop_ids = train_updates[train_updates['pH'].isnull()]['seq_id'].values
train = train[~(train['seq_id'].isin(drop_ids))]
train.shape
```

Les autres 'seq_id' sont accompagnés des modifications à apporter.

```python
change_ids = train_updates[train_updates['pH'].notnull()]['seq_id'].values
```

```python
train_updates = train_updates[train_updates['pH'].notnull()]
```

```python
train.loc[change_ids].head()
```

```python
train_updates.head()
```

Visiblement l'erreur était une inversion des valeurs de 'pH' et 'tm'. Effectuons la correction.

```python
train.loc[train_updates['seq_id'], :] = train_updates.set_index('seq_id')[:]
```

```python
train.loc[change_ids].head()
```

Il nous manque des valeurs de 'seq_id', qui correspondent en fait à l'index.

```python
train['seq_id'] = train['seq_id'].fillna(train.index.to_series())
```

```python
train.loc[change_ids].head()
```

Pour finir on convertit le type de 'seq_id' en entier et on crée un nouveau fichier csv.

```python
train['seq_id'] = train['seq_id'].astype(int)
```

```python
train
```

```python
train.to_csv('Data/train_correct.csv', index=False)
```
